# Consolidando vários arquivos .xlsx em um único DataFrame

##### OBSERVAÇÕES:

Ao setar o path dos arquivos é preciso ter prévio conhecimento das bases para que se consolide de forma correta e se há extenções diferente de planilha. 

**Exemplo: .xlsb

```python
#importando bibliotecas
import os
import pandas as pd

#path - onde estão os arquivos
caminho = 'C:/Users/Lenovo/Desktop/BASE_DADOS/Python/'
```

```python
#criando lista vazia para ser populada
df = []

#looping para obter as pastas na memória
for file in os.listdir(caminho): #lendo a pasta pelo diretório
    if file.endswith('.xlsx'):   #filtrando extensão
        df.append(pd.read_excel(os.path.join(caminho, file))) #concatenando caminho
      
print(len(df)) #retorna a quantidade de arquivos dentro da pasta e que será consolidada

df_principal = pd.concat(df, axis=0)

#visualizando dataframe
display(df_principal)
```
